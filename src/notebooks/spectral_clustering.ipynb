```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer


# Instantiate the clustering model and visualizer
model = SpectralClustering(affinity="nearest_neighbors", assign_labels='cluster_qr', eigen_solver='amg',verbose=True, n_jobs=-1) #
visualizer = KElbowVisualizer(model, k=(30,60))

visualizer.fit(scaled_emnist)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3, min_samples=10, algorithm='brute')
labels = db.fit_predict(scaled_emnist)
```

```python
np.unique(labels)
```

```python
start = time.perf_counter()
clustering = SpectralClustering(n_clusters=45, affinity='nearest_neighbors', assign_labels='cluster_qr', eigen_solver='amg', verbose=True, n_jobs=-1)
clustering.fit(scaled_emnist)
print(time.perf_counter() - start)
```

```python
from sklearn.cluster import KMeans

inertias = []

for i in range(1, 50):
    start = time.perf_counter()
    k_clustering = KMeans(n_clusters=i)
    k_clustering.fit(emnist_preds)
    print("n_cluster=", i, "TIME:", time.perf_counter() - start)
    inertias.append(k_clustering.inertia_)
```

```python
plt.plot(range(1, 50), inertias)
```

```python
for i in range(len(inertias) - 1):
    print(i, 'Spadek:', inertias[i] - inertias[i + 1])
```
