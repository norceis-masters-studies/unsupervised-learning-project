```python
import numpy as np
```

```python
import cv2
```

```python
img = cv2.imread('mnist.png')
```

```python
noise = np.random.uniform(0, 1, size=((32, 32)))
```

```python
#np.random.normal
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(noise, cmap='gray')
```

```python
noise2 = np.where(noise > 0.88, 255, 0)
```

```python
plt.imshow(noise2, cmap='gray')
```

```python
img2 = np.where(img > 127, 255, 0)
```

```python
plt.imshow(img2, cmap='gray')
```

```python
plt.imshow((img2.astype(np.float32) + np.expand_dims(noise2, 2)))
```

```python
noise3 = np.expand_dims(noise2, 2)
```

```python
img3 = np.where(noise3 > 128, ~img, img)
```

```python
plt.imshow(np.where(img3 > 128, 255, 0))
```
